Notebook represent the GNEA analysis or pathway enrichment analysis. 
Purpose of GNEA analysis is to understand what genes (from the path of interest) are influenced (up regilated/down regulated) in interest cohort of pateints ( healthy vs cell patients; tumor cells vs healthy cells, b/n different type cells) 

```r
#instalation of needed library
library(clusterProfiler)
library(ggplot2)
library(tidyverse)
library(DOSE)
library(org.Hs.eg.db)
library(enrichplot)
library(dplyr) 
```

data had been loaded from this webside "https://biostatsquid.com/wp-content/uploads/2023/06/severevshealthy_degresults-1.csv". 
Utilizing the following command on MAC:   curl "https://biostatsquid.com/wp-content/uploads/2023/06/severevshealthy_degresults-1.csv" -o "data.csv"

#### Data is publicaly available from the paper "Low-Avidity CD4+ T Cell Responses to SARS-CoV-2 in Unexposed Individuals and Humans with Severe COVID-19" (acher et all 2020), link to the original paper is here: https://www.sciencedirect.com/science/article/pii/S1074761320305033 

Data includes the comparison in gene expression of severe patients vs unexposed. Data already preprocess adn represents the T- statistics for each gene expression. 


```r
# first read file 

data <- read.csv('data.csv', row.name = 1)
str(data)
```

```r
dim(data)
```

```r
head(data)
# data includes p_value, log2fc -> tgose two columns we will utize for filtering the needed genes
```

```r
filtered <- filter( data, (pval < 0.05 & log2fc >0 ) | (pval < 0.05 & log2fc < 0))
```

```r
dim(filtered)
```

```r
# getting gene set 
#1 lets pull the gene from our dataset first 
genes <- filtered$gene_symbol 
```

```r
#2 pull the genes from the reference database: 
# Utilize the following link https://www.gsea-msigdb.org/gsea/msigdb/collections.jsp -> pullout data from the "KEGG_MEDICUS subset of CP"
# press the "gene symbol" on the left navigation bar and load the dataset in your currnt working directory 

#3 lets take care our ref gene sets 
ref_gene <-read.gmt("c2.cp.kegg_medicus.v2023.2.Hs.symbols.gmt")
```

```r
# as you can see the dataset has about 9000 genes, we dont need all of them, lest filter only genes we have in your dataset
dim(ref_gene)

ref_gene <- ref_gene[ref_gene$gene %in% genes, ]
```

```r
dim(ref_gene)
```

```r
# 4 we can save the ref_data genes in file in rds format 
saveRDS(ref_gene, "KEGG_Medicus.rds")
```

```r
# in our filtered data we have signifcant genes, we will have to sort then into up and down regulated. 
# Thus lets create additinal columns and fill it with categorical values 
filtered <- filtered %>% mutate(de = case_when(
  log2fc > 0 & padj < 0.05 ~ 'Up Regulated',
  log2fc < 0 & padj < 0.05 ~ 'Down Regulated',
  padj > 0.05 ~ 'NO'
))

```

```r
filtered <- filtered[filtered$de != 'NO',]
```

```r
diff_de_data <- split(filtered, filtered$de)
```

```r
str(diff_de_data)
```

```r
# Running the cluster profile 
# First lets read our rds file for ref_gene we saved above. 
ref_bg_genes <- readRDS('KEGG_Medicus.rds')
```

```r
#ref_bg_genes$gene
```

```r
matrix_to_list <- function(pws){
  pws.l <- list()
  for (pw in colnames(pws)) {
    pws.l[[pw]] <- rownames(pws)[as.logical(pws[, pw])]
  }
  return(pws.l)
}
```

```r
result <- lapply(names(diff_de_data),
                 function(x) enricher(gene = diff_de_data[[x]]$gene_symbol,
                                      TERM2GENE = ref_bg_genes))
```

```r
names(result) <- names(diff_de_data)
```

```r
result_df <- lapply(names(result), function(x) rbind(result[[x]]@result))
names(result_df) <- names(result)
result_df <- do.call(rbind, result_df)
head(result_df)
```

```r
result_df <- result_df %>% mutate(minuslog10padj = -log10(p.adjust),
                                  diffexpressed = gsub('\\.GOBP.*$|\\.KEGG.*$|\\.REACTOME.*$', '', rownames(result_df)))

```

```r
# Select only down regulated genes
# down regulated genes in severe exposed to covid patients compare to healthy patients 

down_result_df <- result_df[ result_df$diffexpressed == 'Down Regulated',]

down_resul_df <- down_result_df %>% dplyr::select(!c('minuslog10padj', 'diffexpressed')) 

rownames(down_result_df) <- down_result_df$ID
```

```r
enrichres <- new("enrichResult",
                 readable = FALSE,
                 result = down_result_df,
                 pvalueCutoff = 0.05,
                 pAdjustMethod = "BH",
                 qvalueCutoff = 0.2,
                 organism = "human",
                 ontology = "UNKNOWN",
                 gene = data$gene_symbol, # need genes from original dataset (up, down and no changed one) 
                 keytype = "UNKNOWN",
                 universe = unique(ref_bg_genes$gene), # unique reference genes 
                 gene2Symbol = character(0),
                 geneSets = ref_bg_genes) # data set of reference genes which assosiate gene to the pathway 
class(enrichres)
```

```r
# two paths had been down regulated 
# (what is means that this path is down regulated, but the genes which could be in a path may be suppressors, or translation activators) 
barplot(enrichres, showCategory = 20) 
```

```r
dotplot(enrichres, showCategory = 15) + ggtitle("Severe vs Healthy")
```

```r
#A cnetplot shows the link between genes and pathways, think of this as a network.
cnetplot(enrichres)
```

```r
heatplot(enrichres, showCategory = 5)
```
